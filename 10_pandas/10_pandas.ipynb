## Задание 1

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
```

```python
movies.head()
```

```python
ratings.head()
```

```python
ratings[ratings.rating == 5]
```

```python
agg_rate = ratings[['movieId', 'rating']].groupby('movieId').agg('count')
agg_rate
```

```python
joined = movies.merge(agg_rate, on='movieId', how='left')
joined
```

```python
joined[['title', 'rating']].sort_values(by='rating', ascending=False).head(1)
```

## Задание 2

```python
import pandas as pd
```

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
countries_list = ['Latvia', 'Lithuania', 'Estonia']
categories_list = [4, 12, 21]

filtered_countries = data[(data['country'].isin(countries_list)) & (data['category'].isin(categories_list)) & (data['year'].between(2005, 2010)) & (data['quantity'] >= 0)]
filtered_countries.head()
```

```python
sum_quantity = filtered_countries['quantity'].sum()
sum_quantity
```
