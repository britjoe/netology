```python
import requests
from bs4 import BeautifulSoup
import datetime as dt
import time
import pandas as pd
```

```python
main_link = 'https://habr.com/ru/all/'
```

```python
keywords = ['python', 'парсинг']
```

```python
def get_all_urls(link):
    req = requests.get(link)
    soup = BeautifulSoup(req.text, 'html.parser')
    time.sleep(0.3)
    posts = soup.find_all('article', class_='tm-articles-list__item')
    titles = list(map(lambda x: x.find('h2', class_='tm-article-snippet__title tm-article-snippet__title_h2'),posts))
    page_urls = list(map(lambda y: 'https://habr.com' + y.find('a', class_='tm-article-snippet__title-link').get('href'), titles))
    return page_urls
```

```python
all_urls = get_all_urls(main_link)
```

```python
def get_habr_news(all_links):
    habr_news = pd.DataFrame()
    for link in all_links:
        soup = BeautifulSoup(requests.get(link).text)
        time.sleep(0.3)
        title = soup.find('h1').text
        ##print(title)
        date = pd.to_datetime(soup.find('time').get('title'))
        ##print(date)
        text = soup.find('div', xmlns="http://www.w3.org/1999/xhtml").text
        ##print(text)
        row = {'date': date,
               'title': title,
               'link': link,
               'text': text
              }
        habr_news = pd.concat([habr_news, pd.DataFrame([row])])
    return habr_news
```

```python
hn = get_habr_news(all_urls)
hn
```

```python
hn['lower'] = hn['text'].str.lower()
```

```python
hn_filtered = hn[hn['lower'].str.contains('|'.join(keywords))]
hn_filtered[['date', 'title', 'link', 'text']]
```
